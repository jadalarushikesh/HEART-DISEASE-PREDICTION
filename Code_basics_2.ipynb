```python
df4 = df3.copy()
df4
```

```python
import warnings
warnings.filterwarnings('ignore')

df4 = df3.copy()
df4.ExerciseAngina.replace(
    {
        'N': 0,
        'Y': 1
    },
    inplace=True)

df4.ST_Slope.replace(
    {
        'Down': 1,
        'Flat': 2,
        'Up': 3
    },
    inplace=True)

df4.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True)

df4.head()
```

```python
df5 = pd.get_dummies(df4, drop_first=True)
df5.head()
```

```python
x = df5.drop("HeartDisease",axis='columns')
y = df5.HeartDisease

x.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler =StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=30)
```

```python
print(x_train.shape)

print(x_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()
model_rf.fit(x_train, y_train)
model_rf.score(x_test, y_test)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
PCA = PCA(0.95)
x_pca = PCA.fit_transform(x)
x_pca

# 0.95 means we want to keep 95% of the total information (variance) from the original data.
```

```python
x_pca_train,x_pca_test,y_train,y_test = train_test_split(x_pca,y,test_size = 0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_pca_train,y_train)
model.score(x_pca_test,y_test)
```
